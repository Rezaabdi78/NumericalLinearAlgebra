```python
import scipy.linalg as lg
import numpy as np
```

```python
A = 3*np.random.random((10,10))
print(lg.det(A))
```

```python
A_inv = lg.inv(A)
I = np.dot(A,A_inv)
print(I)
```

```python
b = np.random.random((10,1))
x = lg.solve(A,b)
print(x)
```

```python
lu = lg.lu(A,permute_l=False)
len(lu)
```

```python
P = lu[0]
L = lu[1]
U = lu[2]
```

```python
A1 = np.dot(np.dot(P,L),U)
print(A-A1)
```

```python
y = lg.solve(np.dot(P,L),b)
x1 = lg.solve(U,y)
print(x1-x)
```
